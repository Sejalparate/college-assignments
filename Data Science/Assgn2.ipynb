```python
import numpy as  np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([-3, -2, -1, 0, 1, 2, 3, 4, 5, 6])
```

```python
max = np.abs(x).max()
max
```

```python
max_abs_scaling = x/max
max_abs_scaling
```

```python
plt.plot(x, max_abs_scaling)
```

```python
max = np.max(x)
max
```

```python
min = np.min(x)
min
```

```python
minmax_feature = (x - min)/(max - min)
minmax_feature
```

```python
plt.plot(minmax_feature)
```

```python
mean = np.mean(x)
mean
```

```python
std = np.std(x)
std
```

```python
z_score = (x - mean) / std
z_score
```

```python
plt.plot(x, z_score)
```
