# Import libraries and read images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
img = cv2.imread("images\gojo.png")
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
plt.title('Original')
plt.axis('off');
```

```python
def show_img(ax, img, title=None):
    ax.imshow(img)
    ax.set_title(title)
    ax.axis('off');
```

# Gaussian blur

```python
blur = cv2.blur(img, (15,5))

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], blur, 'Gaussian blur')
```

# Kernel blur

```python
kernel = np.ones((4,4), np.float32) / 25    # Defining kernel to perform convolution
dst = cv2.filter2D(img, -1, kernel)         # Output img will have same depth as input img

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], dst, 'Kernel blur')
```

# Median blur

```python
median = cv2.medianBlur(img, 7)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], median, 'Median Blur')
```

# Grayscale

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], gray_img, 'Gray image')
```

# Sharpening

```python
filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
sharpen_img = cv2.filter2D(img, -1, filter)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], sharpen_img, 'Sharpen image')
```
