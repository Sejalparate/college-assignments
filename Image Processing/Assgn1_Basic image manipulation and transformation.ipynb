# Importing necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images\ocean.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
plt.title('Original')
plt.axis('off');
```

```python
img.shape
```

```python
def show_img(ax, img, title=None):
    ax.imshow(img)
    ax.set_title(title)
    ax.axis('off');
```

# Resize image

```python
r = 300.0 / img.shape[0]
dim = (400, int(img.shape[0] * r))
resized_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
# Interpolation types - INTER_AREA : when shrinking 
#                       INTER_CUBIC : slow but more efficient
#                       INTER _LINEAR : when zooming

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], resized_img, 'Resized')
```

# Rotating image

```python
rotated_img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
# rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
# rotated_img = cv2.rotate(img, cv2.ROTATE_180)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], rotated_img, 'Rotated')
```

# Crop image

```python
crop_img = img[20:330, 100:400]
cv2.imshow("cropped", crop_img)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
show_img(axs[0], img, 'Original')
show_img(axs[1], crop_img, 'Cropped')
```
